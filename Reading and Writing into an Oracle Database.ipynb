```python
#Establish a connection to the database

load_dotenv()
DIALECT = 'oracle'
SQL_DRIVER = 'cx_oracle'
USERNAME = os.getenv('ORACLE_USERNAME')
PASSWORD = os.getenv('ORACLE_PASSWORD')
HOST = os.getenv('HOST')
PORT = os.getenv('PORT') 
SERVICE = os.getenv('SERVICE')

connection_path = DIALECT + '+' + SQL_DRIVER + '://' + USERNAME + ':' + PASSWORD +'@' + HOST + ':' + str(PORT) + '/?service_name=' + SERVICE

engine = create_engine(connection_path)
```

```python
#Bring data in from the database

data = pd.read_sql('select * from account_master where rownum <= 20', engine)
data.head()
```

```python
#Add columns to the data using conditions

def create_columns(df):
    conditions = [
        (df['lcy_opening_bal'] <1000 ),
        (df['lcy_opening_bal'] >= 1000 )]
    
    # create a list of the values we want to assign for each condition
    values = [ 0, 1]

    # create a new column and use np.select to assign values to it using our lists as arguments
    df['indication'] = np.select(conditions, values)
    return df
```

```python
import numpy as np
data = create_columns(data)
```
